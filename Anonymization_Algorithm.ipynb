# BlackMarker

```python
def blackMarker(src, dest):
  new_src = ""
  new_dest = ""
  src_split = src.split(".")
  dest_split = dest.split(".")

  for i in range(4):
    new_src += len(src_split[i])*"0"
    if i != 3:
      new_src += "."

  for i in range(4):
    new_dest += len(dest_split[i])*"0"
    if i != 3:
      new_dest += "."

  # new_src = str.encode(new_src)
  # new_dest = str.encode(new_dest)
  return new_src, new_dest
```

```python
n = len(src_ip)
blackmarker_src = []
blackmarker_dest = []
for i in range(n):
  src, dest = blackMarker(src_ip[i], dest_ip[i])
  blackmarker_src.append(src)
  blackmarker_dest.append(dest)
```

```python
n = data.columns[3]

data[n] = blackmarker_dest

data.head()
```

# Permutation Algorithm

```python
#Permutation Algorithm 
import numpy as np
def permutation(src, dest):
  new_src = ""
  new_dest = ""
  src_split = src.split(".")
  dest_split = dest.split(".")

  new_src_array = np.random.permutation(src_split)
  new_dest_array = np.random.permutation(dest_split)
  
  for i in range(4):
    new_src += new_src_array[i]
    if i != 3:
      new_src += "."

  for i in range(4):
    new_dest += new_dest_array[i]
    if i != 3:
      new_dest += "."

  # new_src = str.encode(new_src)
  # new_dest = str.encode(new_dest)
  return new_src, new_dest

new_src, new_dest = permutation(src_ip[0], dest_ip[0])
print('new src', new_src)
print('new dest', new_dest)
```

```python
n = len(src_ip)
permutation_src = []
permutation_dest = []
for i in range(n):
  src, dest = permutation(src_ip[i], dest_ip[i])
  permutation_src.append(src)
  permutation_dest.append(dest)

print('permutation src', permutation_src[0:100])
print('permutation dest', permutation_dest[0:100])
```

```python
n = data.columns[1]
data[n] = permutation_src
data.head()
```

```python
n = data.columns[3]
data[n] = permutation_dest
data.head()
```

# Truncated Algorithm

```python
import numpy as np

def setToZero(arr, num):
  ind = 3 - (num // 8)
  mod = num % 8
  for i in range(ind+1, 4):
    arr[i] = 0
  mask = ~((1 << mod)-1)
  arr[ind] &= mask
  return arr
def truncation(src, dest, n):
  new_src = ""
  new_dest = ""
  src_split = src.split(".")
  dest_split = dest.split(".")

  src_split = [int(x) for x in src_split]
  dest_split = [int(x) for x in dest_split]
  src_split = setToZero(src_split, n)
  dest_split = setToZero(dest_split, n)
  
  src_split = [str(x) for x in src_split]
  dest_split = [str(x) for x in dest_split]

  for i in range(4):
    new_src += src_split[i]
    if i != 3:
      new_src += "."

  for i in range(4):
    new_dest += dest_split[i]
    if i != 3:
      new_dest += "."

  # new_src = str.encode(new_src)
  # new_dest = str.encode(new_dest)
  return new_src, new_dest

new_src, new_dest = truncation(src_ip[0], dest_ip[0], 12)
print('new src', new_src)
print('new dest', new_dest)
```

```python
n = len(src_ip)
truncated_src = []
truncated_dest = []
for i in range(n):
  src, dest = truncation(src_ip[i], dest_ip[i], 20)
  truncated_src.append(src)
  truncated_dest.append(dest)

print('truncated src', truncated_src[0:100])
print('truncated dest', truncated_dest[0:100])
```

# Reverse Truncation

```python
import numpy as np

def setToZero(arr, num):
  ind = (num // 8)
  mod = num % 8
  for i in range(0, ind):
    arr[i] = 0
  mod = 32 - mod
  mask = ((1 << mod)-1)
  print('mask', mask)
  arr[ind] &= mask
  return arr
  
def rev_truncation(src, dest, n):
  new_src = ""
  new_dest = ""
  src_split = src.split(".")
  dest_split = dest.split(".")

  src_split = [int(x) for x in src_split]
  dest_split = [int(x) for x in dest_split]
  src_split = setToZero(src_split, n)
  dest_split = setToZero(dest_split, n)
  
  src_split = [str(x) for x in src_split]
  dest_split = [str(x) for x in dest_split]

  for i in range(4):
    new_src += src_split[i]
    if i != 3:
      new_src += "."

  for i in range(4):
    new_dest += src_split[i]
    if i != 3:
      new_dest += "."

  # new_src = str.encode(new_src)
  # new_dest = str.encode(new_dest)
  return new_src, new_dest

print('src', src)
print('dest', dest)
new_src, new_dest = rev_truncation(src, dest, 16)
print('new src', new_src)
print('new dest', new_dest)
```

```python
n = len(src_ip)
rev_src = []
rev_dest = []
for i in range(n):
  src, dest = rev_truncation(src, dest, 16)
  rev_src.append(src)
  rev_dest.append(dest)

print('random src', rev_src[0:100])
print('random dest', rev_dest[0:100])
```
